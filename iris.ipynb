```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colors = ['magenta', 'orange', 'springgreen']
sns.set_palette(colors)
# sns.palplot(sns.color_palette()) 
```

```python
iris = sns.load_dataset("iris")
# iris.head()
```

```python
attrs = iris.iloc[:,0:4].values # array of values from the first four columns 
species = iris.iloc[:,4].values # array of values from the species column
```

```python
with plt.style.context("seaborn-whitegrid"):
    plt.figure(figsize = (9,6))
    for i in ('setosa', 'versicolor', 'virginica'):
       plt.hist(attrs[species==i][:,2], bins='auto', alpha=0.3, ec='black', label=i)
    plt.xlabel("Petal Length for different species, cm", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.legend()
    plt.show()
```
